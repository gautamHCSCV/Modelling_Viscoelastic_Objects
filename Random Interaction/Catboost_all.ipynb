```python
import pandas as pd
import numpy as np
import os
import csv
from datetime import datetime
from catboost import CatBoostRegressor
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

```python
def rs_data(x,y,step=10):
    x_train,y_train = [],[]
    for i in range(0,len(y),step):
        x_train.append(x[i])
        y_train.append(y[i])
    return np.array(x_train),np.array(y_train)
```

# Objects-all

```python
path = '/content/drive/MyDrive/Virtual env/multi_finger_random_data/catboost/'
#print(sorted(os.listdir(path)))
for obj in range(1,10):
    with open('result_random_catboost.txt','a') as f:
        f.write(f'\nObject:{obj}\n')
    x1_train = pd.read_csv(path+f'obj{obj}_random_x1_train.csv')
    x2_train = pd.read_csv(path+f'obj{obj}_random_x2_train.csv')
    f1_train = pd.read_csv(path+f'obj{obj}_random_f_train.csv')
    f1_test = pd.read_csv(path+f'obj{obj}_random_f_test.csv')
    x1_test = pd.read_csv(path+f'obj{obj}_random_x1_test.csv')
    x2_test = pd.read_csv(path+f'obj{obj}_random_x2_test.csv')
    X_train = np.concatenate((x1_train,x2_train),axis=1)

    x_test = np.concatenate((x1_test,x2_test),axis=1)
    y_test = np.array(f1_test)

    x_train, y_train = rs_data(X_train,np.array(f1_train))
    print(x_train.shape, y_train.shape, y_test.shape)

    cat = CatBoostRegressor(iterations = 1200, learning_rate=0.01,bagging_temperature=10)
    t1 = datetime.now()
    cat.fit(x_train,y_train)
    t2 = datetime.now()

    y2_pred = cat.predict(x_test)
    print('Root Mean Squared error: {}\n'.format(mean_squared_error(y_test,y2_pred)**0.5))
    print('Time taken:', t2-t1)
    s1 = f'Relative Root Mean Squared error: {(mean_squared_error(y_test,y2_pred)**0.5)/np.mean(y_test)}\nTime taken:{t2-t1}\n'

    errors = []
    for i in range(len(y_test)):
        errors.append(abs(y_test[i]-y2_pred[i]))
    errors = np.array(errors)

    print(' Maximum Error:', max(errors))
    print('Standard deviation:',np.std(errors))
    print('Mean Absolute error:', np.mean(errors/(np.abs(y_test)+0.01)))
    s2 = f'Maximum Error:{max(errors)}\n  Standard deviation:{np.std(errors)}\n  Mean Absolute error:{np.mean(errors/(np.abs(y_test)+0.01))}\n\n'

    with open('result_random_catboost.txt','a') as f:
        f.write(s1)
        f.write(s2)
```
