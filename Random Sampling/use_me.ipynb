```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from datetime import datetime
from sklearn.metrics import mean_squared_error
from math import sqrt
from catboost import CatBoostRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
def data(x,y,step=10):
    x_train,y_train = [],[]
    for i in range(0,len(y),step):
        x_train.append(x[i])
        y_train.append(y[i])
    return np.array(x_train),np.array(y_train)
```

```python
path = '/content/drive/MyDrive/Virtual env/Data_shashi_gautam/'
path1=path+'new_data_velocity/'
for num in range(1,10):
    if num==1 or num==5 or num==9:
        x1_train = pd.read_csv(path+'obj_{}_new/mix/a{}_x1_mixed_train_n.csv'.format(num,num))
        x2_train = pd.read_csv(path+'obj_{}_new/mix/a{}_x2_mixed_train_n.csv'.format(num,num))
        f1_train = pd.read_csv(path+'obj_{}_new/mix/a{}_f_mixed_train_n.csv'.format(num,num))
        f1_test = pd.read_csv(path+'obj_{}_new/mix/a{}_f_mixed_test_n.csv'.format(num,num))
        x1_test = pd.read_csv(path+'obj_{}_new/mix/a{}_x1_mixed_test_n.csv'.format(num,num))
        x2_test = pd.read_csv(path+'obj_{}_new/mix/a{}_x2_mixed_test_n.csv'.format(num,num))
        print(len(x1_train),len(x1_test),len(f1_train),len(f1_test))
        X_train = np.concatenate((x1_train,x2_train),axis=1)
        X_test = np.concatenate((x1_test,x2_test),axis=1)
    else:
        x1_train = pd.read_csv(path1+'obj_{}_new/a{}_x1_mixed_train_n.csv'.format(num,num))
        x2_train = pd.read_csv(path1+'obj_{}_new/a{}_x2_mixed_train_n.csv'.format(num,num))
        f1_train = pd.read_csv(path1+'obj_{}_new/a{}_f_mixed_train_n.csv'.format(num,num))
        f1_test = pd.read_csv(path1+'obj_{}_new/a{}_f_mixed_test_n.csv'.format(num,num))
        x1_test = pd.read_csv(path1+'obj_{}_new/a{}_x1_mixed_test_n.csv'.format(num,num))
        x2_test = pd.read_csv(path1+'obj_{}_new/a{}_x2_mixed_test_n.csv'.format(num,num))
        print(len(x1_train),len(x1_test),len(f1_train),len(f1_test))
        X_train = np.concatenate((x1_train,x2_train),axis=1)
        X_test = np.concatenate((x1_test,x2_test),axis=1)
    
    x_train,x_test,y_train,y_test = np.array(X_train),np.array(X_test),np.array(f1_train),np.array(f1_test)

    x_train,y_train = data(x_train,y_train)
    print(x_train.shape,y_train.shape)
    cat = CatBoostRegressor(iterations = 1200, learning_rate=0.01,bagging_temperature=10)
    t1 = datetime.now()
    cat.fit(x_train,y_train)
    t2 = datetime.now()

    print('Time taken by CATBOOST Model: ',t2-t1)
    y2_pred = cat.predict(x_test)
    print('Root mean square error of CATBOOST Model: ',sqrt(mean_squared_error(y_test,y2_pred)))

    statement1 = f'\n\nObject-{num}\n'
    statement2 = 'time taken by catboost: {} \nRoot Mean Squared error: {}\n'.format((t2-t1)*0.7, mean_squared_error(y_test,y2_pred)**0.5)
    
    file1 = open("RS_mixed.txt","a")
    file1.write(statement1)
    file1.write(statement2)
    file1.close()
```
